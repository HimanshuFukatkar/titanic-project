```python
x=df[['Annual Income (k$)','Spending Score (1-100)']]
x.head()
```

```python
from sklearn.cluster import KMeans
```

```python
#taking 5 cluster
km1=KMeans(n_clusters=5)

#fiting the input data
km1.fit(x)

km1.inertia_
```

```python
#predicting the lables of the input data
y=km1.predict(x)
```

```python
y
```

```python
#adding the.lables to a column named lable
df['label']=y
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
    
#the elbow curve
plt.figure(figsize=(12,6))
#plt.plot(range(1,11,wcss)
plt.plot(range(1,11),wcss,linewidth=2, c='red',marker='8')
plt.xlabel('k value')
#plt.xticks(np.arange(1,11,1))
plt.ylabel('wcss')
plt.show()
```
